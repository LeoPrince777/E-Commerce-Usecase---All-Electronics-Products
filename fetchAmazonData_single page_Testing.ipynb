```python
#sp = BeautifulSoup(response.text, 'lxml')
sp1 = BeautifulSoup(response.content, "html.parser")
```

```python
product_items_dupli = sp1.find_all('div', class_='sg-col-inner')

# Print the total number of products found
print(f"Found {len(product_items_dupli)} products.")
```

```python
# Find all product items
product_items = sp1.find_all('span', class_='a-size-base-plus a-color-base a-text-normal')

# Print the total number of products found
print(f"Found {len(product_items)} products.")
```

```python
import pandas as pd

# List to hold product data
products = []

for product in product_items_dupli:
    try:
        #product_name = sp1.find('span', class_='a-size-medium a-color-base a-text-normal')
        product_name = product.find('span', class_='a-size-base-plus a-color-base a-text-normal').text.strip()
        #print(product_name )
        #if product_name:
            #product_name = product_name.text.strip()
        #else:
            #product_name = None 
       
        # Product price
        #product_price = product.find('span', class_='a-price')
        #product_price = sp1.find_all('span', class_='a-price-whole')
        #sp1.find('span',attrs ={"class":"a-price"}).find('span',attrs ={"class":"a-offscreen"})
        #product_price_whole = product.find("span", attrs={"class":'a-price'}).find("span", attrs={"class": "a-offscreen"}).text.strip()
        
        #product_price_whole =  product.find("span", attrs={"class": "a-offscreen"}).text.strip()
        #print(product_price_whole)
        product_price= product.find('span', class_ ='a-price').find("span", class_ ="a-offscreen").text.strip()
        #product_price = product.find('span', class_ ='a-price').find("span", class_ ="a-offscreen").text.strip().replace(',', '')

        print(product_price)
        #if product_price_whole:
            #product_price_whole = product_price_whole.text
        #else:
            #product_price_whole = None
        
        #product_price = sp1.find('span',attrs ={"class":"a-price"}).find('span',attrs ={"class":"a-offscreen"}).text



    
        # Product description
        #product_description = product.find('ul', class_='a-section a-spacing-medium a-spacing-top-small').text.strip()

        # Product rating
        product_rating = product.find('span', class_='a-icon-alt').text.strip()
        print(product_rating)

        # Product reviews
        #product_reviews = product.find('span', class_='reviews').text.strip()

        # Product image URL
        product_image_url = product.find('img', class_='s-image')['src']
        print(product_image_url)
        # Add the product data to the list
        products.append({
            'Name': product_name,
            'Price': product_price,
            #'Description': product_description,
            'Rating': product_rating,
            #'Reviews': product_reviews,
            'Image_URL': product_image_url
        })

    except AttributeError:
        continue

# Convert the list to a DataFrame for better readability
df = pd.DataFrame(products)
```
